```python
df_raw = pd.read_sql("""SELECT personal_info.uuid,
personal_info.age,
personal_info.dem_has_children,
personal_info.gender,
career_info.dem_education_level,
career_info.dem_full_time_job,
career_info.normalized_job_code,
country_info.country_code,
poll_info.question_bbi_2016wave4_basicincome_argumentsagainst,
poll_info.question_bbi_2016wave4_basicincome_argumentsfor,
poll_info.question_bbi_2016wave4_basicincome_awareness 
FROM personal_info
JOIN career_info 
ON career_info.uuid = personal_info.uuid
JOIN country_info 
ON country_info.uuid = personal_info.uuid
JOIN poll_info 
ON poll_info.uuid = country_info.uuid 
GROUP BY career_info.uuid;""",engine)
df_raw
```

```python
import requests
```

```python
req = requests.get('http://api.dataatwork.org/v1/jobs?')
r_j_son = req.json()
r_j_son = pd.DataFrame(r_j_son)
#r_j_son.head()
```

```python
#dataframe just with jobs and normalized names
jobs_df = r_j_son.drop(columns=['parent_uuid','links'])
jobs_df = jobs_df.rename(columns={'uuid':'normalized_job_code'})
jobs_df.to_csv('jobs_df.csv',index=False)
```
